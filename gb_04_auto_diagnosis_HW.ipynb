## Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import cv2  #opencv
import sys

#from keras.applications.inception_v3 import InceptionV3
from keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator, array_to_img
from keras.applications import InceptionV3
from keras.layers import Dense, GlobalAveragePooling2D
from keras.optimizers import Adam
from keras.models import Model, clone_model

from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report

from pathlib import Path
```

## Data load

**Path to images files**

```python
train_dir=Path('D:/04_Jupyter/HW_04/photo_chest_xray/train')
test_dir=Path('D:/04_Jupyter/HW_04/photo_chest_xray/test')
```

### Training set

**Train data**

```python
normal_dir= train_dir / 'NORMAL'
pneumonia_dir= train_dir / 'PNEUMONIA'

normal_cases=Path(normal_dir).glob('*.jpeg')
pneumonia_cases=Path(pneumonia_dir).glob('*.jpeg')

# getting names and paths of files
train_data_stats=[]

for img in normal_cases:
    train_data_stats.append((str(img),0,'NORMAL'))

for img in pneumonia_cases:
    train_data_stats.append((img, 1,'PNEUMONIA'))

train_data_stats=pd.DataFrame(train_data_stats, columns=['img_path', 'class_ind', 'class_name'],index=None)

Y_train_data_stats=train_data_stats.class_ind

x_train_data, x_val_data, y_train_data, y_val_data = train_test_split( train_data_stats, Y_train_data_stats, test_size=0.2, random_state=42)
x_train_data=x_train_data.reset_index(drop=True) 
x_val_data=x_val_data.reset_index(drop=True) 
y_train_data=y_train_data.reset_index(drop=True) 
y_val_data=y_val_data.reset_index(drop=True) 

display(train_data_stats.shape)
del Y_train_data_stats
```

**Test data**

```python
tst_normal_dir= test_dir / 'NORMAL'
tst_pneumonia_dir= test_dir / 'PNEUMONIA'

tst_normal_cases=Path(tst_normal_dir).glob('*.jpeg')
tst_pneumonia_cases=Path(tst_pneumonia_dir).glob('*.jpeg')

# getting names and paths of files
test_data_stats=[]

for img in tst_normal_cases:
    test_data_stats.append((str(img),0,'NORMAL'))

for img in tst_pneumonia_cases:
    test_data_stats.append((img, 1,'PNEUMONIA'))
    
test_data_stats=pd.DataFrame(test_data_stats, columns=['img_path', 'class_ind', 'class_name'],index=None)
```

### Images count by classes

```python
train_data_cnt=x_train_data[[ 'class_name', 'class_ind']].groupby('class_name').count()

sns.barplot(train_data_cnt.index, train_data_cnt.class_ind)
plt.title('# of cases')
plt.xlabel('Class', fontsize=12)
plt.ylabel('Count', fontsize=12)
train_data_cnt=train_data_cnt.append(pd.DataFrame([train_data_cnt.sum()],index=['Total']))
train_data_cnt['pct_class_ind']=round(train_data_cnt.class_ind/train_data_cnt.class_ind[train_data_cnt.shape[0]-1]*100,2)
display(train_data_cnt)
```

Lowering size of PNEUMONIA samples to balance data.

```python
x_train_data_under=pd.concat( [x_train_data.loc[x_train_data.class_ind==0],
x_train_data.loc[x_train_data.class_ind==1].sample(n=train_data_cnt.loc['NORMAL','class_ind'], random_state=1).reset_index(drop=True)]
)
x_train_data_under=x_train_data_under.sample(frac=1)
#.reset_index(drop=True)
#x_train_data_under.reset_index(drop=True)

train_data_cnt_under=x_train_data_under[[ 'class_name', 'class_ind']].groupby('class_name').count()

sns.barplot(train_data_cnt_under.index, train_data_cnt_under.class_ind)
plt.title('# of cases')
plt.xlabel('Class', fontsize=12)
plt.ylabel('Count', fontsize=12)
train_data_cnt_under=train_data_cnt_under.append(pd.DataFrame([train_data_cnt_under.sum()],index=['Total']))
train_data_cnt_under['pct_class_ind']=round(train_data_cnt_under.class_ind/train_data_cnt_under.class_ind[train_data_cnt_under.shape[0]-1]*100,2)
display(train_data_cnt_under)
```

### Loading images

#### Loading train images

```python
train_data=[]
train_labels=[]
label=0
i=0
j=0

for path in x_train_data.img_path[:]:
    i=i+1
    #display(i)
   
    img = cv2.imread(str(path))
    img = cv2.resize(img, (224,224))
    if img.shape[2] ==1:
        img = np.dstack([img, img, img])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.astype(np.float32)/255.  #scaling pixel values to range (0,1) - for better performance of model
    label =  x_train_data[x_train_data.img_path==path].values[0,1]
    train_data.append(img)
    train_labels.append(label)

    if i%250==0:
        print(i)
         
train_data=np.asarray(train_data)
train_labels=np.asarray(train_labels)
train_labels_cat = to_categorical(train_labels, num_classes=2,dtype='int32')

print('shape of data:',train_data.shape)
print('shape of labels:',train_labels.shape)
```

#### Loading undersampled images

```python
train_data_us=[]
train_labels_us=[]
label=0
i=0
j=0

for path in x_train_data_under.img_path[:]:
    i=i+1
    #display(i)
   
    img = cv2.imread(str(path))
    img = cv2.resize(img, (224,224))
    if img.shape[2] ==1:
        img = np.dstack([img, img, img])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.astype(np.float32)/255.  #scaling pixel values to range (0,1) - for better performance of model
    label =  x_train_data_under[x_train_data_under.img_path==path].values[0,1]
    train_data_us.append(img)
    train_labels_us.append(label)

    if i%250==0:
        print(i)

train_data_us=np.asarray(train_data_us)
train_labels_us=np.asarray(train_labels_us)
train_labels_cat_us = to_categorical(train_labels_us, num_classes=2,dtype='int32')

print('shape of data:',train_data_us.shape)
print('shape of labels:',train_labels_us.shape)
```

```python
val_data=[]
val_labels=[]
label=0
i=0
j=0

for path in x_val_data.img_path[:]:
    i=i+1
    #display(i)
   
    img = cv2.imread(str(path))
    img = cv2.resize(img, (224,224))
    if img.shape[2] ==1:
        img = np.dstack([img, img, img])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.astype(np.float32)/255.  #scaling pixel values to range (0,1) - for better performance of model
    label =  x_val_data[x_val_data.img_path==path].values[0,1]
    val_data.append(img)
    val_labels.append(label)

    if i%250==0:
        print(i)
            
val_data=np.asarray(val_data)
val_labels=np.asarray(val_labels)
val_labels_cat = to_categorical(val_labels, num_classes=2,dtype='int32')

print('shape of data:',val_data.shape)
print('shape of labels:',val_labels.shape)
```

#### Loading test images

```python
test_data=[]
test_labels=[]
label=0
i=0
j=0

for path in test_data_stats.img_path[:]:
    i=i+1
    #display(i)
   
    img = cv2.imread(str(path))
    img = cv2.resize(img, (224,224))
    if img.shape[2] ==1:
        img = np.dstack([img, img, img])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.astype(np.float32)/255.  #scaling pixel values to range (0,1) - for better performance of model
    label =  test_data_stats[test_data_stats.img_path==path].values[0,1]
    test_data.append(img)
    test_labels.append(label)

    if i%250==0:
        print(i)

test_data=np.asarray(test_data)
test_labels=np.asarray(test_labels)
test_labels_cat = to_categorical(test_labels, num_classes=2,dtype='int32')

print('shape of data:',test_data.shape)
print('shape of labels:',test_labels.shape)
```

## Models fit

### Loading Inception V3 model

**Model for further changes**

```python
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

### Adding new to layers

```python
# Add a global spatial average pooling layer
out = base_model.output
out = GlobalAveragePooling2D()(out)
# Add a fully-connected layer 
out = Dense(512, activation='relu')(out)
# Add logistic layer with 2 classes
predictions = Dense(2, activation='softmax')(out) # 2 classes to be predicted
```

### Models to train

```python
# train only the top layers (which were randomly initialized)
# i.e. freeze all convolutional InceptionV3 layers
for layer in base_model.layers:
    layer.trainable = False
```

```python
model_v1 = Model(inputs = base_model.input, outputs = predictions)
model_v2 = clone_model(model_v1)
model_v3 = clone_model(model_v1)
model_v2.set_weights(model_v1.get_weights())
model_v3.set_weights(model_v1.get_weights())

# list layers 
#for i,layer in enumerate(model_v1.layers):
#    print(i,layer.name)
```

### Compiling models

```python
# compile model with optimizer
model_v1.compile(Adam(lr=.0001), loss='binary_crossentropy', metrics=['accuracy'])  # or 'categorical_crossentropy'
model_v2.compile(Adam(lr=.0001), loss='binary_crossentropy', metrics=['accuracy'])  # or 'categorical_crossentropy'
model_v3.compile(Adam(lr=.0001), loss='binary_crossentropy', metrics=['accuracy'])  # or 'categorical_crossentropy'

model_v1.summary()
```

### Fitting the model - try #1- full data
